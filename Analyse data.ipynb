### Giá diện tích theo m2 trung bình theo quận trong tỉnh

```python
# create new table
ads_property_m2=ads_property[['property_id','price_proposed_on_m2']]
property_area=propertys[['id','address_id']]

address_area=address[['id','district_id','province_id']]
address_area.rename(columns={'id':'id_address'}, inplace=True)

district_area=district[['id','name']]
district_area.rename(columns={'id':'id_district','name':'name_district'}, inplace=True)
```

```python
# join
price_m2_ward=pd.merge(property_area,ads_property_m2,
                 left_on='id',right_on='property_id', how='inner')

price_m2_ward=pd.merge(price_m2_ward,address_area,
                 left_on='address_id',right_on='id_address', how='left')

price_m2_ward=pd.merge(price_m2_ward,district_area,
                 left_on='district_id',right_on='id_district', how='left')
```

```python
price_m2_ward.info()
```

```python
# price_m2_ward: bảng đã kết
def plot_price_mean_m2_district(id_province,price_m2_ward):
    #clearning
    price_m2_ward=price_m2_ward.dropna(subset=['price_proposed_on_m2'])
    group=price_m2_ward[price_m2_ward.province_id==id_province].groupby(['id_district'])
    data=[]
    for id_district, g in group:
        name=g['name_district'].unique()[0]
        price_mean=g['price_proposed_on_m2'].mean()
        data.append([id_district,name,price_mean])
    groupby_district=pd.DataFrame(columns=['id_district','name_district','price_proposed_on_m2_mean'],data=data)
    
    plt.figure(figsize=(25,10))
    plt.bar(groupby_district.name_district, groupby_district.price_proposed_on_m2_mean)
    plt.show()
```

```python
plot_price_mean_m2_district(249,price_m2_ward)
```
